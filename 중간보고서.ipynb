### 데이터 전처리 및 정제

```python
import pandas as pd
import numpy as np
```

```python
subway = pd.read_csv('Documents/data/subway.csv', encoding = 'cp949')
subway.head()   #상위 5개 데이터 확인. (2021년 4월까지 수집되어있음을 확인.)
```

```python
#지하철 2호선 데이터만 추출.

subway02 = subway[subway.호선명 == '2호선']
subway02
```

```python
# 2호선 지하철역 정보 확인.

set(subway02.지하철역)
```

```python
subway.info()   #데이터 타입 확인, 결측치 확인.
```

```python
#사용월 데이터타입 변경.

from datetime import datetime
subway02['사용월'] = pd.to_datetime(subway02['사용월'],format = '%Y%m') 
subway02
```

```python
#필요없는 열 제거 (작업일자,호선명)

subway02 = subway02.drop( columns = ['작업일자','호선명'])  #2호선만 모아있는 상태이므로 '호선명' 열제거.
subway02.head(20)
```

```python
#열이름에 변수가 포함되어 있으므로 melt함수 활용하여 분할.

subway02_melt = subway02.melt(id_vars=['사용월','지하철역'],
                var_name = 'time_type',
                value_name = 'total')

subway02_melt.head(20)
```

```python
subway02_melt = ( subway02.melt(id_vars=['사용월','지하철역'],
                var_name = 'time_type',
                value_name = 'total')
                ['time_type']
                .str.split(expand = True)
                .rename(columns = {0:'time', 1:'type'})
                .assign(type = lambda x : x.type.str[0:2]    )              #승차인원 -> 승차로 변경. (type)
                        #time = lambda y : y.time.str[0:2]+y.time.str[3:6])   # 04시-05시 ->04-05로 변경(time)
                )

subway02_melt
```

```python
subway02_melt = subway02.melt(id_vars=['사용월','지하철역'],
                var_name = 'time_type',
                value_name = 'total')

time_type = (subway02_melt['time_type']
            .str.split(expand = True)
            .rename(columns = {0:'time', 1:'type'})
            .assign(type = lambda x : x.type.str[0:2],                    #승차인원 -> 승차로 변경. (type)
                    time = lambda y : y.time.str[0:2]+y.time.str[3:6]))   # 04시-05시 ->04-05로 변경(time))

subway02_concat=pd.concat([subway02_melt[['사용월','지하철역','total']], time_type],axis =1)
subway02_concat
```


### EDA



```python
# 지하철역별 평균 승객수 내림차순으로 정렬.

subway02_mean = (
    subway02_concat
    .groupby(['지하철역'])['total']
    .agg(['mean'])
    .sort_values('mean',ascending=False) 
)
subway02_mean = subway02_mean[['mean']].astype(int)   #평균값 정수로 변환.
subway02_mean
```

```python
 #2015 ~2021년 4월 지하철 2호선 평균 승객수.
subway02_mean.mean().astype(int)  
```
